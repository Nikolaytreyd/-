### Задача 5
Значения в разных «столбцах» таблицы с данными об эмодзи имеют разный масштаб — например, в Твиттере он значительно больше. Чтобы оценить эту разницу, посчитайте среднее количество эмодзи из первой двадцатки для EmojiXpress, Инстаграма и Твиттера. 
Оформим вывод красиво (Это помогает сделать выводы быстрее)

```python
from prettytable import PrettyTable

def get_mean_id(id:int=10) -> str:
    table = PrettyTable()
    table.field_names = ['НАЗВАНИЕ', f'Среднее за {id}']
    table.add_row(['EMOJIXPRESS', f"{data['EMOJIXPRESS, МЛН'].iloc[id].mean()} млн."])
    table.add_row(['INSTAGRAM', f"{data['INSTAGRAM, МЛН'].iloc[id].mean()} млн."])
    table.add_row(['ТВИТТЕР', f"{data['ТВИТТЕР, МЛН'].iloc[id].mean()} млн."])
    print(table)

get_mean_id()
```
