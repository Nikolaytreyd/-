```python
import requests

def get_exchange_rate():
    url = 'https://api.exchangerate-api.com/v4/latest/USD'
    response = requests.get(url)
    data = response.json()
    rate = data['rates']['RUB']
    return rate

# Пример использования функции
exchange_rate = get_exchange_rate()
print(f"Курс рубля к доллару: {exchange_rate}")
```

```python
def round_mod(x, option=1):
    return round(x, option)

data['Бюджет, руб'] = data['Бюджет, $'] * exchange_rate
data['Сборы, руб'] = data['Сборы, $'] * exchange_rate

data['Бюджет, руб'] = data['Бюджет, руб'].apply(round_mod)
data['Сборы, руб'] = data['Сборы, руб'].apply(round_mod)

data
```

```python
class General:
    def __init__(self, data, option_rub=True):
        self.data = data

    def create_profit(self, option_rub=True):
        if option_rub:
            self.data['Прибыль, руб'] = self.data['Сборы, руб'] - self.data['Бюджет, руб']
            return self.data
        else:
            self.data['Прибыль, $'] = self.data['Сборы, $'] - self.data['Бюджет, $']
            return self.data

    def sort_mod(self, option_rub=True, option_revers=False, head=5, name='Прибыль, руб'):
        self.create_profit(option_rub)
        if (
                option_rub and name == 'Прибыль, $'
                or not option_rub and name == 'Прибыль, руб'

        ):
            raise NameError(f'Невозможно провести сортировку по ## {name} ##, так как option_rub = {option_rub}')
        else:
            new_data = data.sort_values(by=name, ascending=option_revers)
            return new_data.iloc[0:head]




run = General(data=data)
```
